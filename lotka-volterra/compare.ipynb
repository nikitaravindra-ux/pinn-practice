```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# load classical solution
t           = np.load("ode_t.npy")
ode_rabbits = np.load("ode_rabbits.npy")
ode_foxes   = np.load("ode_foxes.npy")
```

```python
# load PINN predictions
t_pinn       = np.load("pinn_t.npy")
pinn_rabbits = np.load("pinn_rabbits.npy")
pinn_foxes   = np.load("pinn_foxes.npy")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Lotka-Volterra: PINN vs scipy odeint", fontsize=14, fontweight='bold')

#rabbits
axes[0].plot(t, ode_rabbits, 'b-',  linewidth=2, label='scipy odeint')
axes[0].plot(t_pinn, pinn_rabbits, 'r--', linewidth=2, label='PINN')
axes[0].set_title('Rabbits (Prey)')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Population')
axes[0].legend()
axes[0].grid(True)

#foxes
axes[1].plot(t, ode_foxes, 'b-',  linewidth=2, label='scipy odeint')
axes[1].plot(t_pinn, pinn_foxes, 'r--', linewidth=2, label='PINN')
axes[1].set_title('Foxes (Predator)')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Population')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig('comparison_plot.png', dpi=150) #saves img
plt.show()
```
